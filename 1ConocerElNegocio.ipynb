# **Problemas de Negocio que Aborda el Dataset de Ventas de Tacos**

1. fluctuación en la Demanda:
Uno de los principales problemas de la industria gastronómica, especialmente en la comida rápida, es la fluctuación en la demanda. Esto puede generar desperdicios de comida o falta de insumos. El dataset puede ayudar a analizar las variaciones de demanda en diferentes días, semanas o meses, y así poder prever con más precisión la cantidad de ingredientes necesarios para cada día.

2. optimización de Inventarios:
Sin un sistema adecuado de gestión de inventarios, los negocios pueden enfrentar pérdidas de productos perecederos como tortillas, carne, salsa, etc. Con los datos del dataset, los dueños de los restaurantes pueden optimizar el control de insumos, reduciendo desperdicios y mejorando la rentabilidad.

3. identificación de Tendencias de Consumo:
Este dataset ofrece una visión importante sobre qué tipo de tacos tienen mayor demanda, en qué zonas geográficas se consume más y en qué días de la semana se concentran los pedidos. Esto ayuda a las marcas a realizar una planificación adecuada de su producción y estrategia de marketing.

# **Oportunidades para el Negocio de Tacos en Chile**

Expansión en Nuevas Zonas Geográficas:
El análisis de las zonas con mayor demanda puede revelar nuevas oportunidades de expansión. Un negocio exitoso de tacos podría considerar abrir nuevas ubicaciones en zonas con alta densidad poblacional o expandirse a ciudades fuera de Santiago, como Viña del Mar o Concepción, donde la comida mexicana comienza a ser más popular.

Promociones Específicas:
Con la información sobre los días de mayor y menor demanda, las marcas pueden diseñar ofertas y promociones para los días más tranquilos, como el lunes o martes, aumentando las ventas en días de baja demanda.

Personalización de Menú:
Adaptarse a las preferencias locales también es una forma de innovar. Ofrecer tacos con ingredientes que resuenen con los gustos de los chilenos, como tacos de pescado frito, carne de cordero o incluso tacos con productos típicos como avellanas y frutillas puede atraer a más clientes.


## 📊 La Parte Operacional Del Negocio


# **Este modelo Pipeline  mostrata el conjunto de procesos o etapas de este Negocio**

```python

import matplotlib.pyplot as plt
import matplotlib.patches as patches

# --- Datos del pipeline ---
data_pipeline = {
    'Captura del Dato': [
        'Ubicación del cliente',
        'Tipo y tamaño de taco',
        'Cantidad de toppings',
        'Tiempo de entrega',
        'Distancia recorrida',
        'Propina entregada',
        'Patrones de día'
    ],
    'Análisis Aplicado': [
        'Segmentación geográfica',
        'Análisis de preferencias',
        'Control de costos',
        'Evaluación de eficiencia',
        'Optimización de rutas',
        'Medición de satisfacción',
        'Análisis comparativo'
    ],
    'Decisión Estratégica': [
        'Planificación de marketing',
        'Ingeniería de menú',
        'Ajuste de precios',
        'Mejora de operaciones',
        'Reducción de gastos',
        'Programas de lealtad',
        'Gestión de demanda'
    ]
}

# --- Configuración del gráfico ---
fig, ax = plt.subplots(figsize=(18, 8))
fig.patch.set_facecolor('#F1F3F5')  # Fondo general suave

# Colores para etapas
etapa_colores = ['#0D47A1', '#1976D2', '#42A5F5']  # Gama profesional azul

# Posiciones horizontales de las etapas
x_pos = [0.15, 0.5, 0.85]
etapas = list(data_pipeline.keys())
y_pos = 0.5

# Estilo de las cajas
box_style = dict(boxstyle='round,pad=0.5', fc='white', ec='#CED4DA', lw=1.8)
arrow_style = dict(arrowstyle='-|>', color='#343A40', lw=2, mutation_scale=18)

# Dibujar cajas y títulos
for i, etapa in enumerate(etapas):
    x = x_pos[i]
    items = '\n'.join(f'• {line}' for line in data_pipeline[etapa])

    # Texto de la etapa
    ax.text(x, y_pos, items,
            ha='center', va='center',
            fontsize=11, color='#495057',
            bbox=box_style, linespacing=1.5)

    # Título superior de cada etapa
    ax.text(x, 0.9, f'{i+1}. {etapa}',
            ha='center', va='center',
            fontsize=15, fontweight='bold',
            color=etapa_colores[i])

# Flechas entre etapas
for i in range(len(x_pos) - 1):
    arrow = patches.FancyArrowPatch(
        (x_pos[i] + 0.13, y_pos),
        (x_pos[i+1] - 0.13, y_pos),
        connectionstyle="arc3,rad=0.2",
        **arrow_style
    )
    ax.add_patch(arrow)

# Formato final
ax.axis('off')
ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
plt.suptitle('🔍 Pipeline de Inteligencia de Negocio – Taco Sales',
             fontsize=22, fontweight='bold', color='#212529', y=0.98)
plt.tight_layout(rect=[0, 0, 1, 0.94])
plt.show()
```

```python

import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd

eventos = {
    "Fecha": [
        "2020-01-01", "2021-06-15", "2022-03-10",
        "2023-08-05", "2024-01-01", "2025-06-01"
    ],
    "Evento": [
        "Idea inicial del negocio: tacos personalizados con modelo basado en datos.",
        "Primera sucursal en Austin, TX: validación de mercado.",
        "Sistema de pedidos con captura de datos por cliente.",
        "Expansión a Dallas, San Antonio y New York.",
        "Aplicación de analítica para optimizar logística.",
        "Evaluación del mercado chileno para replicar el modelo."
    ]
}

df_eventos = pd.DataFrame(eventos)
df_eventos["Fecha"] = pd.to_datetime(df_eventos["Fecha"])

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(df_eventos["Fecha"], [1]*len(df_eventos), "o", color="saddlebrown")

for i, row in df_eventos.iterrows():
    ax.plot([row["Fecha"], row["Fecha"]], [1, 1.05], color="gray", linestyle="--")
    ax.text(row["Fecha"], 1.07, row["Evento"], ha='center', va='bottom', rotation=45, fontsize=9)

ax.get_yaxis().set_visible(False)
ax.set_title("TACO", fontsize=14, fontweight='bold')
ax.xaxis.set_major_locator(mdates.YearLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
